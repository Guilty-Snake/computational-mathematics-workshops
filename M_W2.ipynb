```python
#Qn.1
def distinctelement(list):
  #variable for the unique elements
  unique = []
  for i in list:
    if i not in unique:
      #adding the unique elements
      unique.append(i)
  return unique
sample = [1,2,3,3,3,3,4,5]
print(distinctelement(sample))
```

```python
#Qn.2
def ifprime(n):
  #since 1 is not prime
  if n == 1:
    return False
  #initializing loop to check prime number
  for i in range(2,n):
    if n%i == 0:
      return False
    else:
      return True
print(ifprime(11))
```

```python

#Qn.3
#funtion to check whether a number is prime
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print("Twin primes less than 1000:")
#loop to check if two consecutive odd numbers are both prime i.e. twin prime
for i in range(3, 1000, 2):  # checks odd numbers only
    if is_prime(i) and is_prime(i + 2):
        print(f"({i}, {i+2})")
```

```python
#Qn. 4
def sumOfDivisors(n):
    total = 0
    #Loop to check if 'n' is divisible by other numbers excluding itself
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total

# Example
print(sumOfDivisors(6))  # Output: 6
```

```python
#Qn.5
#Reusing the function from Qn.4
def sumOfDivisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total
#perfect number +1 to find the next perfect number
n = 29

while True:
    if sumOfDivisors(n) == n:
        print("Next perfect number is:", n)
        break
    n += 1
```

```python
#Qn.6
Marks = [20, 30, 35, 55, 70, 80, 90]

# Mean
total = 0
for num in Marks:
    total += num
mean = total / len(Marks)

# Median
Marks.sort()
n = len(Marks)

if n % 2 == 1:
    median = Marks[n // 2]
else:
    median = (Marks[n//2 - 1] + Marks[n//2]) / 2

print("Mean:", mean)
print("Median:", median)
```

```python
#Qn.7
def decimalToBinary(n):
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n = n // 2
    return binary

print(decimalToBinary(10))  # Output: 1010
```

```python
#Qn.8
def prodDigits(n):
    product = 1
    while n > 0:
        digit = n % 10
        product *= digit
        n = n // 10
    return product

print(prodDigits(234))  # Output: 24
```

```python
#Qn.9
def multiply_all(*args):
    result = 1
    for num in args:
        result *= num
    return result

# a. 1,2,3,4,5
print("Result (1,2,3,4,5):", multiply_all(1,2,3,4,5))

# b. 12,13,14
print("Result (12,13,14):", multiply_all(12,13,14))
```

```python
#set operations
# Define the sets
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}
set3 = {10, 20, 30}

# a.
# len() returns the number of elements in the set
print("Length of set1:", len(set1))


# b.
# manually adding elements using a loop
total = 0
for num in set2:
    total += num
print("Sum of set2:", total)


# c.
# union returns all unique elements from both sets
print("Union:", set1 | set2)


# d.
# intersection returns common elements in both sets
print("Intersection:", set1 & set2)


# e.
# elements present in set1 but not in set2
print("set1 - set2:", set1 - set2)


# f.
# elements present in set2 but not in set1
print("set2 - set1:", set2 - set1)


# g.
# issubset() checks if all elements of set3 are in another set
print("set3 subset of set1:", set3.issubset(set1))
print("set3 subset of set2:", set3.issubset(set2))


# h.
# complement = elements in the universal set but not in set1
U = set(range(1, 51))
print("Complement of set1:", U - set1)


# i.
# 'in' keyword checks membership
print("15 in set1:", 15 in set1)


# j.
# add() inserts a new element into the set
set1.add(60)
print("Updated set1:", set1)


# k.
# discard() removes element if present
set2.discard(60)
print("Updated set2:", set2)


# l.
# clear() empties the entire set
set3.clear()
print("set3 after clearing:", set3)


# m.
# max() returns largest value, min() returns smallest value
print("Max of set2:", max(set2))
print("Min of set2:", min(set2))
```
